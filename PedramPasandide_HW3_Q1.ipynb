```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating noisy sinusoidal data
start = np.pi/4
end = 2.5*np.pi
X = np.linspace(start,end, 42)
y = np.sin(X)

y_noisy = y + np.random.normal(0, 0.5, size=int(len(X)))

e1= X[int(len(X)/3)]
e2 = X[int(2*len(X)/3)]
```

```python
fig, ax = plt.subplots()

ax.scatter(X,y_noisy, edgecolors='black')
ax.plot(X,y)

plt.hlines(y=1, xmin=X[0], xmax=e1, color='g')
plt.hlines(y=-0.75, xmin=e1, xmax=e2, color='g')
plt.hlines(y=0.5, xmin=e2, xmax=X[-1], color='g')
ax.axvline(e1)
ax.axvline(e2)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(X,y_noisy, edgecolors='black')
ax.plot(X,y)


ax.plot([1, e1], [1.5, 0.3],color='g')
ax.plot([e1, e2], [-0.6, -1],color='g')
ax.plot([e2, X[-1]], [-0.5, 1],color='g')
ax.axvline(e1)
ax.axvline(e2)

plt.show()
```

```python
xe1 = np.linspace(start, e1, int(42/3)) 
xe1 = xe1 + np.random.normal(0,0.05, size=int(len(xe1)))
ye1 = 0*xe1+0

xe1_end = np.linspace(e1, end, int(42/3*2)) + np.random.normal(0,0.05, size=int(42/3*2))
ye1_end = 3*xe1_end + -3*e1
fig, ax = plt.subplots()

ax.scatter(xe1,ye1,c='black')
ax.scatter(xe1_end,ye1_end,c='black')


ax.plot([start, e1], [ye1[0], ye1[-1]],color='g')
ax.plot([e1,X[-1]],[ye1_end[0],ye1_end[-1]],color='g')
ax.axvline(e1)
ax.axvline(e2)

plt.show()
```
